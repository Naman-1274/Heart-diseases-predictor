```python
df.duplicated().sum()
data = df.drop_duplicates()
```

```python
numeric_feature = [feature for feature in df.columns if data[feature].dtype != 'O']
print('we have {} no of numeric_features are : {}'.format(len(numeric_feature),numeric_feature))
```

```python
categorical_feature = [feature for feature in df.columns if data[feature].dtypes == 'O']
print('we have {} no of categorical_feature are : {}'.format(len(categorical_feature),categorical_feature))
```

```python
numeric_features = ['HeartDiseaseorAttack','HighBP','HighChol','BMI','Smoker','Diabetes','PhysActivity','Fruits','Veggies','HvyAlcoholConsump','MentHlth','PhysHlth','Sex','Age','Education','Income']
for feature in numeric_features:
    sns.histplot(data[feature],kde=True)
    plt.title(f'Distribution of {feature}')
    plt.figure(figsize=(10, 5))
    plt.show()
```

```python
Matrix = data.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(Matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(data,hue='HeartDiseaseorAttack',vars = numeric_features)
plt.show()
```

```python
data['HighBP_HighChol'] = data['HighBP'] & data['HighChol']
data['Smoker_Diabetes'] = data['Smoker'] & data['Diabetes']
data['Smoker_HvyAlcoholConsump'] = data['Smoker'] & data['HvyAlcoholConsump']
data['BMI_PhysActivity'] = data['BMI'] & data['PhysActivity']
```
